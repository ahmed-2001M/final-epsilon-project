## test scraping data from wuzzuf and save them in temp file and load it again to use it in deployment

```python
wuzzuf = WUZZUF()
wuzzuf.wuzzufRunner(job_name='data analysis',n_scraping_pages=1,dbg=False)
skills = wuzzuf.get_skills()
```

```python
unique_skills = set()
for i in skills:
    for skill in i:
        unique_skills.add(skill)
unique_skills = list(unique_skills)
```

```python
import pickle
```

```python
with open("ts.pkl", "wb") as file:
    pickle.dump(unique_skills, file)
```
